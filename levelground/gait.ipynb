```python
import pandas as pd

# Load each CSV file and display the first few rows to understand the data
file_paths = [
    'LFL.csv',
    'LSL.csv',
    'LTL.csv',
    'PW.csv',
    'RFL.csv',
    'RSL.csv',
    'RTL.csv'
]

# 데이터프레임을 로드하고 'Series0' 열을 삭제
dataframes = {}
for file_path in file_paths:
    # 데이터프레임 로드
    df = pd.read_csv(file_path)
    df.drop(index=0, inplace=True)
    df.drop(index=1, inplace=True)

    # 'Series0' 열이 있는 경우, 해당 열 삭제
if 'Series0' in df.columns:
    df.drop(columns=['Series0'], inplace=True)
    
    # 수정된 데이터프레임을 사전에 저장
dataframes[file_path.split('/')[-1].split('.')[0]] = df

round_df = df.round(2)

round_df


```
